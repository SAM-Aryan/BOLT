### Importing the Libraries

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score
from catboost import CatBoostRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

### Importing the dataset

```python
dataset1 = pd.read_csv("https://raw.githubusercontent.com/vujicictijana/MIDAS/main/Dataset/IceCreamFactory_100_Ramp.csv")
dataset2 = pd.read_csv("https://raw.githubusercontent.com/vujicictijana/MIDAS/main/Dataset/IceCreamFactory_102_Step.csv")
dataset3 = pd.read_csv("https://raw.githubusercontent.com/vujicictijana/MIDAS/main/Dataset/IceCreamFactory_105_Freeze.csv")
```

### Taking care of the missing data

#### For dataset 1


```python
to_be_deleted = []
for i in list(dataset1.columns) :
    if len(dataset1[i].unique()) == 1 :
        to_be_deleted.append(i)

dataset1.drop(to_be_deleted, axis=1)
```

#### For dataset 2

```python
to_be_deleted = []
for i in list(dataset2.columns) :
    if len(dataset2[i].unique()) == 1 :
        to_be_deleted.append(i)

dataset2.drop(to_be_deleted, axis=1)
```

#### For dataset 3

```python
to_be_deleted = []
for i in list(dataset3.columns) :
    if len(dataset3[i].unique()) == 1 :
        to_be_deleted.append(i)

dataset3.drop(to_be_deleted, axis=1)
```

### Mixing the three datasets into one

```python
dataset12 = pd.merge(dataset1, dataset2, how="outer")
dataset = pd.merge(dataset12, dataset3, how="outer")
dataset = dataset.drop(["Parameter for Anomaly", "Timestamp"], axis = 1)
```

### Dividing the data into features and targets

```python
X  = dataset.drop("Anomaly", axis = 1)
y = dataset["Anomaly"]
print(y)
```

### Splitting the data into test and train set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

## LightGBM

```python
lightgbm_model = LGBMRegressor(n_estimators=100, max_depth=3, learning_rate=0.1, objective='regression')
lightgbm_model.fit(X_train, y_train)
y_pred_lightgbm = lightgbm_model.predict(X_test)
```

## XGBoost

```python
xg_boost_model = XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.1, objective='reg:squarederror')
xg_boost_model.fit(X_train._get_numeric_data(), np.ravel(y_train, order='C')).predict(X_test._get_numeric_data())
y_pred_xg_boost = xg_boost_model.predict(X_test)
```

## CatBoost

```python
cat_boost_model = CatBoostRegressor(learning_rate=1, depth=8, loss_function="RMSE")
cat_boost_model.fit(X_train, y_train)
y_pred_cat_boost = cat_boost_model.predict(X_test)
```
